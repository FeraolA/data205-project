## Clean Traffic Violations

```python
tv = tv_raw.copy()

# Column names
tv.columns = tv.columns.str.strip().str.lower().str.replace(' ', '_', regex=False)

# Parse dates and times
tv['date_of_stop'] = pd.to_datetime(tv['date_of_stop'], errors='coerce')
tv['time_of_stop'] = pd.to_datetime(tv['time_of_stop'], format='%H:%M:%S', errors='coerce').dt.time

# Temporal features
tv['year']        = tv['date_of_stop'].dt.year
tv['month']       = tv['date_of_stop'].dt.month
tv['month_name']  = tv['date_of_stop'].dt.month_name()
tv['day_of_week'] = tv['date_of_stop'].dt.day_name()
tv['hour']        = pd.to_datetime(tv['time_of_stop'].astype(str), format='%H:%M:%S', errors='coerce').dt.hour
tv['time_period'] = pd.cut(
    tv['hour'],
    bins=[-1, 4, 11, 16, 20, 23],
    labels=['Night', 'Morning', 'Afternoon', 'Evening', 'Night2']
)
tv['time_period'] = tv['time_period'].astype(str).replace('Night2', 'Night')

# Yes/No → 1/0
bool_cols = ['accident', 'belts', 'personal_injury', 'property_damage',
             'fatal', 'commercial_license', 'hazmat', 'commercial_vehicle',
             'alcohol', 'work_zone', 'search_conducted', 'contributed_to_accident']
for col in [c for c in bool_cols if c in tv.columns]:
    tv[col] = tv[col].map({'Yes': 1, 'No': 0})

# Standardize categoricals
for col in ['violation_type', 'race', 'gender', 'arrest_type', 'search_outcome']:
    if col in tv.columns:
        tv[col] = tv[col].astype(str).str.strip().str.title().replace({'Nan': np.nan})

# Fix bad coordinates
for col in ['latitude', 'longitude']:
    if col in tv.columns:
        tv[col] = pd.to_numeric(tv[col], errors='coerce')
        tv.loc[tv[col] == 0, col] = np.nan

# Filter
tv = tv.dropna(subset=['date_of_stop'])
tv = tv[tv['year'] >= 2018].reset_index(drop=True)
```

## Clean Parking Inventory

```python
park = park_raw.copy()
park.columns = park.columns.str.strip().str.lower().str.replace(' ', '_', regex=False)
park = park.dropna(subset=['location']) if 'location' in park.columns else park
```
